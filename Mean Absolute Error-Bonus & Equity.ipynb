```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
fpath = 'salaries_modified.csv'
fpath2 = 'salary.csv'
```

```python
df_cleaned = pd.read_csv(fpath)
df = pd.read_csv(fpath2)
df
```

```python
bonus_mean = df_cleaned.bonus.mean()
equity_mean = df_cleaned.stockgrantvalue.mean()
```

```python
df['b_mean'] = bonus_mean
df['e_mean'] = equity_mean
df.head()
```

```python
b_y_true = df.bonus
b_y_pred = df.b_mean
e_y_true = df.stockgrantvalue
e_y_pred = df.e_mean
```

```python
b_MAE = mean_absolute_error(b_y_true,b_y_pred)
b_MAE
```

```python
e_MAE = mean_absolute_error(e_y_true,e_y_pred)
e_MAE
```
